Prompt 15: Trend of rides per day

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df.groupby(df['date'].dt.date)['booking_id'].count().plot(figsize=(12,5))
plt.ylabel("Number of Rides")
plt.title("Daily Ride Count")
plt.show()
```

Prompt 16: Peak booking times

```python
df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S', errors='coerce')
df['hour'] = df['time'].dt.hour
df['hour'].value_counts().sort_index().plot(kind='bar', figsize=(12,5))
plt.xlabel("Hour of Day")
plt.ylabel("Number of Bookings")
plt.title("Peak Booking Hours")
plt.show()
```
